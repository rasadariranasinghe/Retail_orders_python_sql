```python
import pandas as pd
```

```python
df = pd.read_csv('orders.csv')
df.head(20)
```

```python
df['Ship Mode'].unique()
```

```python
# handling null values
new_df = df.copy()
new_df = pd.read_csv('orders.csv', na_values= ['Not Available', 'unknown'])
new_df['Ship Mode'].unique()
```

```python
new_df.columns = new_df.columns.str.lower()
new_df.columns = new_df.columns.str.replace(' ', '_')
new_df.head()
```

```python
new_df['sales_price'] = ((new_df['list_price']) * (100 - (new_df['discount_percent'])))/100
new_df.head()
```

```python
new_df ["profit"] = new_df ['sales_price'] - new_df ['cost_price']
new_df.head()
```

```python
# dropping unnecessary columns
neww_df = new_df .copy()
neww_df = neww_df.drop(columns = ['cost_price','list_price','discount_percent'])
neww_df.head()
```

```python
neww_df.dtypes
```

```python
neww_df["order_date"]= pd.to_datetime(neww_df["order_date"],format = "%Y-%m-%d")
```

```python
neww_df.dtypes
```

```python
# load the data into sql server
import sqlalchemy as sal
```

```python
# Define connection parameters
username = "postgres"
password = "postgres"
host = "localhost" 
port = "5432"  
database = "retail"

# Create the PostgreSQL connection string
connection_string = f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"

# Create the engine
engine = sal.create_engine(connection_string)

# Connect to the database
conn = engine.connect()
print("Connected to PostgreSQL successfully!")
```

```python
# Write DataFrame to SQL
neww_df.to_sql('orders', con=engine, index=False, if_exists='append')

print("Data inserted into PostgreSQL successfully!")
```
